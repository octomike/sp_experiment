# Initial analysis

```python
sub = '3'
```

```python
fname = 'sub-{}_task-spactive_events.tsv'.format(sub)
fpath = op.join(data_dir, fname)

# Read the data
df = pd.read_csv(fpath, sep='\t')
```

```python
%%timeit
ntrials = int(df['trial'].max()+1)

arr = []
for number in range(1, 10):
    counter0 = 0
    counter1 = 0
    for trial in range(ntrials):
        payoff_dict = get_payoff_dict(df, trial)
        appended_distributions = [i for j in payoff_dict.values() for i in j]
        if number not in appended_distributions:
            continue
        
        # Which option was the number in
        opt = int(appended_distributions.index(number) >= len(payoff_dict[0]))
        
        # Was that option finally selected?
        selected_option = df[df['action_type'] == 'final_choice']['action'].tolist()[trial]
        if opt == selected_option:
            counter1 += 1

        # Increment counter of generally number encountered
        counter0 += 1
    
    # Add proportion of this number
    arr.append(counter1/counter0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,10), arr)
```

```python
plt.figure()
for sub in range(1, 5):
    
    fname = 'sub-{}_task-spactive_events.tsv'.format(sub)
    fpath = op.join(data_dir, fname)

    # Read the data
    df = pd.read_csv(fpath, sep='\t')
    
    
    ntrials = int(df['trial'].max()+1)

    arr = []
    for number in range(1, 10):
        counter0 = 0
        counter1 = 0
        for trial in range(ntrials):
            payoff_dict = get_payoff_dict(df, trial)
            appended_distributions = [i for j in payoff_dict.values() for i in j]
            if number not in appended_distributions:
                continue

            # Which option was the number in
            opt = int(appended_distributions.index(number) >= len(payoff_dict[0]))

            # Was that option finally selected?
            selected_option = df[df['action_type'] == 'final_choice']['action'].tolist()[trial]
            if opt == selected_option:
                counter1 += 1

            # Increment counter of generally number encountered
            counter0 += 1

        # Add proportion of this number
        arr.append(counter1/counter0)
        
    
    plt.plot(range(1,10), arr, label=sub)
    
plt.legend()
```
